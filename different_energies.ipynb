```python
import ROOT
from ROOT import TFile, TCanvas, TLegend, TLatex
```

```python

_700_all_file_name="~/test/R3BRoot/build/bin/700_all.root"
_700_all_file=TFile(_700_all_file_name,"read")
_700_all_file_hist_monitor=_700_all_file.GetDirectory("HistMonitor")
_700_all_hit_energy_hist=_700_all_file_hist_monitor.Get("hE")
_700_protons_file_name="~/test/R3BRoot/build/bin/700_protons.root"
_700_protons_file=TFile(_700_protons_file_name,"read")
_700_protons_file_hist_monitor=_700_protons_file.GetDirectory("HistMonitor")
_700_protons_hit_energy_hist=_700_protons_file_hist_monitor.Get("hE")
_700_no_electrons_positrons_gamma_file_name="~/test/R3BRoot/build/bin/700_no_electrons_positrons_gamma.root"
_700_no_electrons_positrons_gamma_file=TFile(_700_no_electrons_positrons_gamma_file_name,"read")
_700_no_electrons_positrons_gamma_file_hist_monitor=_700_no_electrons_positrons_gamma_file.GetDirectory("HistMonitor")
_700_no_electrons_positrons_gamma_hit_energy_hist=_700_no_electrons_positrons_gamma_file_hist_monitor.Get("hE")

```

```python
%jsroot off
canvas=TCanvas("hE","Hit Energy",0,0,2000,2000)
canvas.Divide(2,2)
canvas.cd(1)
_500_all_hit_energy_hist.SetTitle("n=1, E=500MeV")
_500_protons_hit_energy_hist.SetTitle("n=1, E=500MeV")
_500_no_electrons_positrons_gamma_hit_energy_hist.SetTitle("n=1, E=500MeV")


_500_all_hit_energy_hist.Draw()
_500_all_hit_energy_hist.SetLineColor(4)
_500_protons_hit_energy_hist.Draw("same hist")
_500_protons_hit_energy_hist.SetLineColor(2)
_500_no_electrons_positrons_gamma_hit_energy_hist.Draw("same hist")
_500_no_electrons_positrons_gamma_hit_energy_hist.SetLineColor(6)

_500_all_hit_energy_hist.Fit("gaus","0","",10.,20.)
_500_protons_hit_energy_hist.Fit("gaus","0","",10.,20.)
_500_no_electrons_positrons_gamma_hit_energy_hist.Fit("gaus","0","",10.,20.)

_500_all_hit_energy_hist.SetStats(0)
_500_protons_hit_energy_hist.SetStats(0)
_500_no_electrons_positrons_gamma_hit_energy_hist.SetStats(0)

_500_legend=TLegend(0.6,0.7,0.9,0.9)
_500_legend.SetHeader("Hit Energies", "c")
_500_legend.AddEntry(_500_all_hit_energy_hist,"all particles","l")
_500_legend.AddEntry(_500_protons_hit_energy_hist,"only protons","l")
_500_legend.AddEntry(_500_no_electrons_positrons_gamma_hit_energy_hist,"no e⁺,e⁻,γ","l")
_500_legend.Draw()

_500_all_hit_energy_hist.GetYaxis().SetRangeUser(0,1400)

canvas.cd(2)
_700_all_hit_energy_hist.SetTitle("n=1, E=700MeV")
_700_protons_hit_energy_hist.SetTitle("n=1, E=700MeV")
_700_no_electrons_positrons_gamma_hit_energy_hist.SetTitle("n=1, E=700MeV")


_700_all_hit_energy_hist.Draw()
_700_all_hit_energy_hist.SetLineColor(4)
_700_protons_hit_energy_hist.Draw("same hist")
_700_protons_hit_energy_hist.SetLineColor(2)
_700_no_electrons_positrons_gamma_hit_energy_hist.Draw("same hist")
_700_no_electrons_positrons_gamma_hit_energy_hist.SetLineColor(6)

_700_all_hit_energy_hist.Fit("gaus","0","",10.,20.)
_700_protons_hit_energy_hist.Fit("gaus","0","",10.,20.)
_700_no_electrons_positrons_gamma_hit_energy_hist.Fit("gaus","0","",10.,20.)

_700_all_hit_energy_hist.SetStats(0)
_700_protons_hit_energy_hist.SetStats(0)
_700_no_electrons_positrons_gamma_hit_energy_hist.SetStats(0)

_700_legend=TLegend(0.6,0.7,0.9,0.9)
_700_legend.SetHeader("Hit Energies", "c")
_700_legend.AddEntry(_700_all_hit_energy_hist,"all particles","l")
_700_legend.AddEntry(_700_protons_hit_energy_hist,"only protons","l")
_700_legend.AddEntry(_700_no_electrons_positrons_gamma_hit_energy_hist,"no e⁺,e⁻,γ","l")
_700_legend.Draw()

_700_all_hit_energy_hist.GetYaxis().SetRangeUser(0,1600)

canvas.cd(3)
_1000_all_hit_energy_hist.SetTitle("n=1, E=1000MeV")
_1000_protons_hit_energy_hist.SetTitle("n=1, E=1000MeV")
_1000_no_electrons_positrons_gamma_hit_energy_hist.SetTitle("n=1, E=1000MeV")


_1000_all_hit_energy_hist.Draw()
_1000_all_hit_energy_hist.SetLineColor(4)
_1000_protons_hit_energy_hist.Draw("same hist")
_1000_protons_hit_energy_hist.SetLineColor(2)
_1000_no_electrons_positrons_gamma_hit_energy_hist.Draw("same hist")
_1000_no_electrons_positrons_gamma_hit_energy_hist.SetLineColor(6)

_1000_all_hit_energy_hist.Fit("gaus","0","",8.,18.)
_1000_protons_hit_energy_hist.Fit("gaus","0","",8.,18.)
_1000_no_electrons_positrons_gamma_hit_energy_hist.Fit("gaus","0","",8.,18.)

_1000_all_hit_energy_hist.SetStats(0)
_1000_protons_hit_energy_hist.SetStats(0)
_1000_no_electrons_positrons_gamma_hit_energy_hist.SetStats(0)

_1000_legend=TLegend(0.6,0.7,0.9,0.9)
_1000_legend.SetHeader("Hit Energies", "c")
_1000_legend.AddEntry(_1000_all_hit_energy_hist,"all particles","l")
_1000_legend.AddEntry(_1000_protons_hit_energy_hist,"only protons","l")
_1000_legend.AddEntry(_1000_no_electrons_positrons_gamma_hit_energy_hist,"no e⁺,e⁻,γ","l")
_1000_legend.Draw()
_1000_all_hit_energy_hist.GetYaxis().SetRangeUser(0,2200)

canvas.cd(4)
_1300_all_hit_energy_hist.SetTitle("n=1, E=1300MeV")
_1300_protons_hit_energy_hist.SetTitle("n=1, E=1300MeV")
_1300_no_electrons_positrons_gamma_hit_energy_hist.SetTitle("n=1, E=1300MeV")


_1300_all_hit_energy_hist.Draw()
_1300_all_hit_energy_hist.SetLineColor(4)
_1300_protons_hit_energy_hist.Draw("same hist")
_1300_protons_hit_energy_hist.SetLineColor(2)
_1300_no_electrons_positrons_gamma_hit_energy_hist.Draw("same hist")
_1300_no_electrons_positrons_gamma_hit_energy_hist.SetLineColor(6)

_1300_all_hit_energy_hist.Fit("gaus","","same",6.,17.)
_1300_protons_hit_energy_hist.Fit("gaus","0","",6.,17.)
_1300_no_electrons_positrons_gamma_hit_energy_hist.Fit("gaus","0","",6.,17.)

_1300_all_hit_energy_hist.SetStats(0)
_1300_protons_hit_energy_hist.SetStats(0)
_1300_no_electrons_positrons_gamma_hit_energy_hist.SetStats(0)

_1300_legend=TLegend(0.6,0.7,0.9,0.9)
_1300_legend.SetHeader("Hit Energies", "c")
_1300_legend.AddEntry(_1300_all_hit_energy_hist,"all particles","l")
_1300_legend.AddEntry(_1300_protons_hit_energy_hist,"only protons","l")
_1300_legend.AddEntry(_1300_no_electrons_positrons_gamma_hit_energy_hist,"no e+,e-,","l")
_1300_legend.Draw()

_1300_all_hit_energy_hist.GetYaxis().SetRangeUser(0,2500)

canvas.Draw()
canvas.SaveAs("energies.png")
```
