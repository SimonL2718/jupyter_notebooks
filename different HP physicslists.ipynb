```python
import ROOT, math
from ROOT import TFile, TCanvas, TLegend, TLatex
```

```python
root_file_folder_path="~/test/data/simu_data/physicslists/"
```

```python
file_BERT_HP=TFile(root_file_folder_path+"all_particles_above_0MeV_1000MeV_n1_QGSP_BERT_HP.root","read")
hist_hE_BERT_HP=file_BERT_HP.GetDirectory("HistMonitor").Get("hE")

file_BIC_HP=TFile(root_file_folder_path+"all_particles_above_0MeV_1000MeV_n1_QGSP_BIC_HP.root","read")
hist_hE_BIC_HP=file_BIC_HP.GetDirectory("HistMonitor").Get("hE")

file_INCLXX_HP=TFile(root_file_folder_path+"all_particles_above_0MeV_1000MeV_n1_QGSP_INCLXX_HP.root","read")
hist_hE_INCLXX_HP=file_INCLXX_HP.GetDirectory("HistMonitor").Get("hE")
```

```python
#We want to scale the histograms so they all have a value of 1000 at the main peak of the data.

#set accordingly to cut out the first big peak in the data according to bin number
lower_peak_bin=42

#set the value the peaks should be normalised to
normalized_peak_height=1000
```

```python
hist_hE_BERT_HP.GetXaxis().SetRange(lower_peak_bin,hist_hE_BERT_HP.GetNbinsX())
scaled_hist_hE_BERT_HP=normalized_peak_height/hist_hE_BERT_HP.GetBinContent(hist_hE_BERT_HP.GetMaximumBin())*hist_hE_BERT_HP

hist_hE_BIC_HP.GetXaxis().SetRange(lower_peak_bin,hist_hE_BIC_HP.GetNbinsX())
scaled_hist_hE_BIC_HP=normalized_peak_height/hist_hE_BIC_HP.GetBinContent(hist_hE_BIC_HP.GetMaximumBin())*hist_hE_BIC_HP

hist_hE_INCLXX_HP.GetXaxis().SetRange(lower_peak_bin,hist_hE_INCLXX_HP.GetNbinsX())
scaled_hist_hE_INCLXX_HP=normalized_peak_height/hist_hE_INCLXX_HP.GetBinContent(hist_hE_INCLXX_HP.GetMaximumBin())*hist_hE_INCLXX_HP
```

```python
%jsroot on
canvas=TCanvas("hE","Hit Energy",0,0,1000,1000)
```

```python
scaled_hist_hE_aa_title="Hit Energies"
scaled_hist_hE_BERT_HP.SetTitle(scaled_hist_hE_aa_title)
scaled_hist_hE_BIC_HP.SetTitle(scaled_hist_hE_aa_title)
scaled_hist_hE_INCLXX_HP.SetTitle(scaled_hist_hE_aa_title)

scaled_hist_hE_BERT_HP.SetLineColor(4)
scaled_hist_hE_BIC_HP.SetLineColor(2)
scaled_hist_hE_INCLXX_HP.SetLineColor(6)

scaled_hist_hE_BERT_HP.Draw("same hist")
scaled_hist_hE_BIC_HP.Draw("same hist")
scaled_hist_hE_INCLXX_HP.Draw("same hist")

scaled_hist_hE_BERT_HP.SetStats(0)
scaled_hist_hE_BIC_HP.SetStats(0)
scaled_hist_hE_INCLXX_HP.SetStats(0)

legend_aa=TLegend(0.6,0.7,0.9,0.9)

legend_aa.AddEntry(scaled_hist_hE_BERT_HP,"BERT_HP")
legend_aa.AddEntry(scaled_hist_hE_BIC_HP,"BIC_HP")
legend_aa.AddEntry(scaled_hist_hE_INCLXX_HP,"INCLXX_HP")
legend_aa.SetHeader("Hit Energies","C")
legend_aa.Draw()


scaled_hist_hE_BERT_HP.GetXaxis().SetRange(0,scaled_hist_hE_BERT_HP.GetNbinsX())
scaled_hist_hE_BIC_HP.GetXaxis().SetRange(0,scaled_hist_hE_BIC_HP.GetNbinsX())
scaled_hist_hE_INCLXX_HP.GetXaxis().SetRange(0,scaled_hist_hE_INCLXX_HP.GetNbinsX())
hist_aa_global_max_list=[scaled_hist_hE_BERT_HP.GetBinContent(scaled_hist_hE_BERT_HP.GetMaximumBin()),
                         scaled_hist_hE_BIC_HP.GetBinContent(scaled_hist_hE_BIC_HP.GetMaximumBin()),
                         scaled_hist_hE_INCLXX_HP.GetBinContent(scaled_hist_hE_INCLXX_HP.GetMaximumBin())]
scaled_hist_hE_BERT_HP.GetYaxis().SetRangeUser(0,math.floor(max(hist_aa_global_max_list)/100)*100+100)
scaled_hist_hE_BIC_HP.GetYaxis().SetRangeUser(0,math.floor(max(hist_aa_global_max_list)/100)*100+100)
scaled_hist_hE_INCLXX_HP.GetYaxis().SetRangeUser(0,math.floor(max(hist_aa_global_max_list)/100)*100+100)
```

```python
canvas.Draw()
canvas.SaveAs("different HP physicslists.png")
```

```python
#y-value cut-off factor for fits
fit_selection_scale=0.9
fit_selection_value=normalized_peak_height*fit_selection_scale
```

```python
scaled_hist_hE_BERT_HP.Fit("gaus","S 0","same",
                         scaled_hist_hE_BERT_HP.GetXaxis().GetBinLowEdge(
                             scaled_hist_hE_BERT_HP.FindFirstBinAbove(
                                 fit_selection_value,1,lower_peak_bin,scaled_hist_hE_BERT_HP.GetNbinsX())),
                         scaled_hist_hE_BERT_HP.GetXaxis().GetBinUpEdge(
                             scaled_hist_hE_BERT_HP.FindLastBinAbove(
                                 fit_selection_value,1,lower_peak_bin,scaled_hist_hE_BERT_HP.GetNbinsX())))
scaled_hist_hE_BIC_HP.Fit("gaus","S 0","same",
                        scaled_hist_hE_BIC_HP.GetXaxis().GetBinLowEdge(
                            scaled_hist_hE_BIC_HP.FindFirstBinAbove(
                                fit_selection_value,1,lower_peak_bin,scaled_hist_hE_BIC_HP.GetNbinsX())),
                        scaled_hist_hE_BIC_HP.GetXaxis().GetBinUpEdge(
                            scaled_hist_hE_BIC_HP.FindLastBinAbove(
                                fit_selection_value,1,lower_peak_bin,scaled_hist_hE_BIC_HP.GetNbinsX())))
scaled_hist_hE_INCLXX_HP.Fit("gaus","S 0","same",
                        scaled_hist_hE_INCLXX_HP.GetXaxis().GetBinLowEdge(
                            scaled_hist_hE_INCLXX_HP.FindFirstBinAbove(
                                fit_selection_value,1,lower_peak_bin,scaled_hist_hE_INCLXX_HP.GetNbinsX())),
                        scaled_hist_hE_INCLXX_HP.GetXaxis().GetBinUpEdge(
                            scaled_hist_hE_INCLXX_HP.FindLastBinAbove(
                                fit_selection_value,1,lower_peak_bin,scaled_hist_hE_INCLXX_HP.GetNbinsX())))
```

```python
peak_position_hE_BERT_HP=scaled_hist_hE_BERT_HP.GetFunction("gaus").GetParameter("Mean")
peak_position_hE_BIC_HP=scaled_hist_hE_BIC_HP.GetFunction("gaus").GetParameter("Mean")
peak_position_hE_INCLXX_HP=scaled_hist_hE_INCLXX_HP.GetFunction("gaus").GetParameter("Mean")

array_peak_positions_aa=[peak_position_hE_BERT_HP,
                         peak_position_hE_BIC_HP,
                         peak_position_hE_INCLXX_HP]
array_peak_positions_aa
```

```python
template_root_file=TFile("template.root","RECREATE")

template_root_file.WriteObject(scaled_hist_hE_BERT_HP,"BERT_HP_HitEnergy_Hist")
template_root_file.WriteObject(scaled_hist_hE_BIC_HP,"BIC_HP_HitEnergy_Hist")
template_root_file.WriteObject(scaled_hist_hE_INCLXX_HP,"INCLXX_HP_HitEnergy_Hist")

template_root_file.Close()
```
