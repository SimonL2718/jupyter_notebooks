```python
import ROOT, math
from ROOT import TFile, TCanvas, TLegend, TLatex
```

```python
root_file_folder_path="~/test/data/simu_data/physicslists/"
```

```python
file_QGSP_BERT_=TFile(root_file_folder_path+"all_particles_above_0MeV_1000MeV_n1_QGSP_BERT.root","read")
hist_hE_QGSP_BERT_=file_QGSP_BERT_.GetDirectory("HistMonitor").Get("hE")

file_QGSP_BERT_HP=TFile(root_file_folder_path+"all_particles_above_0MeV_1000MeV_n1_QGSP_BERT_HP.root","read")
hist_hE_QGSP_BERT_HP=file_QGSP_BERT_HP.GetDirectory("HistMonitor").Get("hE")

file_QGSP_BERT_EMV=TFile(root_file_folder_path+"all_particles_above_0MeV_1000MeV_n1_QGSP_BERT_EMV.root","read")
hist_hE_QGSP_BERT_EMV=file_QGSP_BERT_EMV.GetDirectory("HistMonitor").Get("hE")
```

```python
file_QGSP_BIC_=TFile(root_file_folder_path+"all_particles_above_0MeV_1000MeV_n1_QGSP_BIC.root","read")
hist_hE_QGSP_BIC_=file_QGSP_BIC_.GetDirectory("HistMonitor").Get("hE")

file_QGSP_BIC_HP=TFile(root_file_folder_path+"all_particles_above_0MeV_1000MeV_n1_QGSP_BIC_HP.root","read")
hist_hE_QGSP_BIC_HP=file_QGSP_BIC_HP.GetDirectory("HistMonitor").Get("hE")

file_QGSP_BIC_EMV=TFile(root_file_folder_path+"all_particles_above_0MeV_1000MeV_n1_QGSP_BIC_EMV.root","read")
hist_hE_QGSP_BIC_EMV=file_QGSP_BIC_EMV.GetDirectory("HistMonitor").Get("hE")
```

```python
#We want to scale the histograms so they all have a value of 1000 at the main peak of the data.

#set accordingly to cut out the first big peak in the data according to bin number
lower_peak_bin=42

#set the value the peaks should be normalised to
normalized_peak_height=1000
```

```python
hist_hE_QGSP_BERT_.GetXaxis().SetRange(lower_peak_bin,hist_hE_QGSP_BERT_.GetNbinsX())
scaled_hist_hE_QGSP_BERT_=normalized_peak_height/hist_hE_QGSP_BERT_.GetBinContent(hist_hE_QGSP_BERT_.GetMaximumBin())*hist_hE_QGSP_BERT_

hist_hE_QGSP_BERT_HP.GetXaxis().SetRange(lower_peak_bin,hist_hE_QGSP_BERT_HP.GetNbinsX())
scaled_hist_hE_QGSP_BERT_HP=normalized_peak_height/hist_hE_QGSP_BERT_HP.GetBinContent(hist_hE_QGSP_BERT_HP.GetMaximumBin())*hist_hE_QGSP_BERT_HP

hist_hE_QGSP_BERT_EMV.GetXaxis().SetRange(lower_peak_bin,hist_hE_QGSP_BERT_EMV.GetNbinsX())
scaled_hist_hE_QGSP_BERT_EMV=normalized_peak_height/hist_hE_QGSP_BERT_EMV.GetBinContent(hist_hE_QGSP_BERT_EMV.GetMaximumBin())*hist_hE_QGSP_BERT_EMV
```

```python
hist_hE_QGSP_BIC_.GetXaxis().SetRange(lower_peak_bin,hist_hE_QGSP_BIC_.GetNbinsX())
scaled_hist_hE_QGSP_BIC_=normalized_peak_height/hist_hE_QGSP_BIC_.GetBinContent(hist_hE_QGSP_BIC_.GetMaximumBin())*hist_hE_QGSP_BIC_

hist_hE_QGSP_BIC_HP.GetXaxis().SetRange(lower_peak_bin,hist_hE_QGSP_BIC_HP.GetNbinsX())
scaled_hist_hE_QGSP_BIC_HP=normalized_peak_height/hist_hE_QGSP_BIC_HP.GetBinContent(hist_hE_QGSP_BIC_HP.GetMaximumBin())*hist_hE_QGSP_BIC_HP

hist_hE_QGSP_BIC_EMV.GetXaxis().SetRange(lower_peak_bin,hist_hE_QGSP_BIC_EMV.GetNbinsX())
scaled_hist_hE_QGSP_BIC_EMV=normalized_peak_height/hist_hE_QGSP_BIC_EMV.GetBinContent(hist_hE_QGSP_BIC_EMV.GetMaximumBin())*hist_hE_QGSP_BIC_EMV
```

```python
%jsroot on
canvas=TCanvas("hE","Hit Energy",0,0,2000,1000)
canvas.Divide(2,1)
```

```python
canvas.cd(1)
scaled_hist_hE_QGSP_BERT_title="QGSP_BERT"
scaled_hist_hE_QGSP_BERT_.SetTitle(scaled_hist_hE_QGSP_BERT_title)
scaled_hist_hE_QGSP_BERT_HP.SetTitle(scaled_hist_hE_QGSP_BERT_title)
scaled_hist_hE_QGSP_BERT_EMV.SetTitle(scaled_hist_hE_QGSP_BERT_title)

scaled_hist_hE_QGSP_BERT_.SetLineColor(4)
scaled_hist_hE_QGSP_BERT_HP.SetLineColor(2)
scaled_hist_hE_QGSP_BERT_EMV.SetLineColor(6)

scaled_hist_hE_QGSP_BERT_.Draw("same hist")
scaled_hist_hE_QGSP_BERT_HP.Draw("same hist")
scaled_hist_hE_QGSP_BERT_EMV.Draw("same hist")

scaled_hist_hE_QGSP_BERT_.SetStats(0)
scaled_hist_hE_QGSP_BERT_HP.SetStats(0)
scaled_hist_hE_QGSP_BERT_EMV.SetStats(0)

legend_QGSP_BERT=TLegend(0.6,0.7,0.9,0.9)

legend_QGSP_BERT.AddEntry(scaled_hist_hE_QGSP_BERT_,"QGSP_BERT")
legend_QGSP_BERT.AddEntry(scaled_hist_hE_QGSP_BERT_HP,"QGSP_BERT_HP")
legend_QGSP_BERT.AddEntry(scaled_hist_hE_QGSP_BERT_EMV,"QGSP_BERT_EMV")
legend_QGSP_BERT.SetHeader("Hit Energies","C")
legend_QGSP_BERT.Draw()


scaled_hist_hE_QGSP_BERT_.GetXaxis().SetRange(0,scaled_hist_hE_QGSP_BERT_.GetNbinsX())
scaled_hist_hE_QGSP_BERT_HP.GetXaxis().SetRange(0,scaled_hist_hE_QGSP_BERT_HP.GetNbinsX())
scaled_hist_hE_QGSP_BERT_EMV.GetXaxis().SetRange(0,scaled_hist_hE_QGSP_BERT_EMV.GetNbinsX())
hist_QGSP_BERT_global_max_list=[scaled_hist_hE_QGSP_BERT_.GetBinContent(scaled_hist_hE_QGSP_BERT_.GetMaximumBin()),
                         scaled_hist_hE_QGSP_BERT_HP.GetBinContent(scaled_hist_hE_QGSP_BERT_HP.GetMaximumBin()),
                         scaled_hist_hE_QGSP_BERT_EMV.GetBinContent(scaled_hist_hE_QGSP_BERT_EMV.GetMaximumBin())]
scaled_hist_hE_QGSP_BERT_.GetYaxis().SetRangeUser(0,math.floor(max(hist_QGSP_BERT_global_max_list)/100)*100+100)
scaled_hist_hE_QGSP_BERT_HP.GetYaxis().SetRangeUser(0,math.floor(max(hist_QGSP_BERT_global_max_list)/100)*100+100)
scaled_hist_hE_QGSP_BERT_EMV.GetYaxis().SetRangeUser(0,math.floor(max(hist_QGSP_BERT_global_max_list)/100)*100+100)
```

```python
canvas.cd(2)

scaled_hist_hE_QGSP_BIC_title="QGSP_BIC"
scaled_hist_hE_QGSP_BIC_.SetTitle(scaled_hist_hE_QGSP_BIC_title)
scaled_hist_hE_QGSP_BIC_HP.SetTitle(scaled_hist_hE_QGSP_BIC_title)
scaled_hist_hE_QGSP_BIC_EMV.SetTitle(scaled_hist_hE_QGSP_BIC_title)

scaled_hist_hE_QGSP_BIC_.SetLineColor(4)
scaled_hist_hE_QGSP_BIC_HP.SetLineColor(2)
scaled_hist_hE_QGSP_BIC_EMV.SetLineColor(6)

scaled_hist_hE_QGSP_BIC_.Draw("same hist")
scaled_hist_hE_QGSP_BIC_HP.Draw("same hist")
scaled_hist_hE_QGSP_BIC_EMV.Draw("same hist")

scaled_hist_hE_QGSP_BIC_.SetStats(0)
scaled_hist_hE_QGSP_BIC_HP.SetStats(0)
scaled_hist_hE_QGSP_BIC_EMV.SetStats(0)

legend_QGSP_BIC=TLegend(0.6,0.7,0.9,0.9)

legend_QGSP_BIC.AddEntry(scaled_hist_hE_QGSP_BIC_,"QGSP_BIC")
legend_QGSP_BIC.AddEntry(scaled_hist_hE_QGSP_BIC_HP,"QGSP_BIC_HP")
legend_QGSP_BIC.AddEntry(scaled_hist_hE_QGSP_BIC_EMV,"QGSP_BIC_EMV")
legend_QGSP_BIC.SetHeader("Hit Energies","C")
legend_QGSP_BIC.Draw()


scaled_hist_hE_QGSP_BIC_.GetXaxis().SetRange(0,scaled_hist_hE_QGSP_BIC_.GetNbinsX())
scaled_hist_hE_QGSP_BIC_HP.GetXaxis().SetRange(0,scaled_hist_hE_QGSP_BIC_HP.GetNbinsX())
scaled_hist_hE_QGSP_BIC_EMV.GetXaxis().SetRange(0,scaled_hist_hE_QGSP_BIC_EMV.GetNbinsX())
hist_QGSP_BIC_global_max_list=[scaled_hist_hE_QGSP_BIC_.GetBinContent(scaled_hist_hE_QGSP_BIC_.GetMaximumBin()),
                         scaled_hist_hE_QGSP_BIC_HP.GetBinContent(scaled_hist_hE_QGSP_BIC_HP.GetMaximumBin()),
                         scaled_hist_hE_QGSP_BIC_EMV.GetBinContent(scaled_hist_hE_QGSP_BIC_EMV.GetMaximumBin())]
scaled_hist_hE_QGSP_BIC_.GetYaxis().SetRangeUser(0,math.floor(max(hist_QGSP_BIC_global_max_list)/100)*100+100)
scaled_hist_hE_QGSP_BIC_HP.GetYaxis().SetRangeUser(0,math.floor(max(hist_QGSP_BIC_global_max_list)/100)*100+100)
scaled_hist_hE_QGSP_BIC_EMV.GetYaxis().SetRangeUser(0,math.floor(max(hist_QGSP_BIC_global_max_list)/100)*100+100)
```

```python
canvas.Draw()
canvas.SaveAs("different_physicslists.png")
```

```python
#y-value cut-off factor for fits
fit_selection_scale=0.9
fit_selection_value=normalized_peak_height*fit_selection_scale
```

```python
scaled_hist_hE_QGSP_BERT_.Fit("gaus","S 0","same",
                         scaled_hist_hE_QGSP_BERT_.GetXaxis().GetBinLowEdge(
                             scaled_hist_hE_QGSP_BERT_.FindFirstBinAbove(
                                 fit_selection_value,1,lower_peak_bin,scaled_hist_hE_QGSP_BERT_.GetNbinsX())),
                         scaled_hist_hE_QGSP_BERT_.GetXaxis().GetBinUpEdge(
                             scaled_hist_hE_QGSP_BERT_.FindLastBinAbove(
                                 fit_selection_value,1,lower_peak_bin,scaled_hist_hE_QGSP_BERT_.GetNbinsX())))
scaled_hist_hE_QGSP_BERT_HP.Fit("gaus","S 0","same",
                        scaled_hist_hE_QGSP_BERT_HP.GetXaxis().GetBinLowEdge(
                            scaled_hist_hE_QGSP_BERT_HP.FindFirstBinAbove(
                                fit_selection_value,1,lower_peak_bin,scaled_hist_hE_QGSP_BERT_HP.GetNbinsX())),
                        scaled_hist_hE_QGSP_BERT_HP.GetXaxis().GetBinUpEdge(
                            scaled_hist_hE_QGSP_BERT_HP.FindLastBinAbove(
                                fit_selection_value,1,lower_peak_bin,scaled_hist_hE_QGSP_BERT_HP.GetNbinsX())))
scaled_hist_hE_QGSP_BERT_EMV.Fit("gaus","S 0","same",
                        scaled_hist_hE_QGSP_BERT_EMV.GetXaxis().GetBinLowEdge(
                            scaled_hist_hE_QGSP_BERT_EMV.FindFirstBinAbove(
                                fit_selection_value,1,lower_peak_bin,scaled_hist_hE_QGSP_BERT_EMV.GetNbinsX())),
                        scaled_hist_hE_QGSP_BERT_EMV.GetXaxis().GetBinUpEdge(
                            scaled_hist_hE_QGSP_BERT_EMV.FindLastBinAbove(
                                fit_selection_value,1,lower_peak_bin,scaled_hist_hE_QGSP_BERT_EMV.GetNbinsX())))
```

```python
scaled_hist_hE_QGSP_BIC_.Fit("gaus","S 0","same",
                         scaled_hist_hE_QGSP_BIC_.GetXaxis().GetBinLowEdge(
                             scaled_hist_hE_QGSP_BIC_.FindFirstBinAbove(
                                 fit_selection_value,1,lower_peak_bin,scaled_hist_hE_QGSP_BIC_.GetNbinsX())),
                         scaled_hist_hE_QGSP_BIC_.GetXaxis().GetBinUpEdge(
                             scaled_hist_hE_QGSP_BIC_.FindLastBinAbove(
                                 fit_selection_value,1,lower_peak_bin,scaled_hist_hE_QGSP_BIC_.GetNbinsX())))
scaled_hist_hE_QGSP_BIC_HP.Fit("gaus","S 0","same",
                        scaled_hist_hE_QGSP_BIC_HP.GetXaxis().GetBinLowEdge(
                            scaled_hist_hE_QGSP_BIC_HP.FindFirstBinAbove(
                                fit_selection_value,1,lower_peak_bin,scaled_hist_hE_QGSP_BIC_HP.GetNbinsX())),
                        scaled_hist_hE_QGSP_BIC_HP.GetXaxis().GetBinUpEdge(
                            scaled_hist_hE_QGSP_BIC_HP.FindLastBinAbove(
                                fit_selection_value,1,lower_peak_bin,scaled_hist_hE_QGSP_BIC_HP.GetNbinsX())))
scaled_hist_hE_QGSP_BIC_EMV.Fit("gaus","S 0","same",
                        scaled_hist_hE_QGSP_BIC_EMV.GetXaxis().GetBinLowEdge(
                            scaled_hist_hE_QGSP_BIC_EMV.FindFirstBinAbove(
                                fit_selection_value,1,lower_peak_bin,scaled_hist_hE_QGSP_BIC_EMV.GetNbinsX())),
                        scaled_hist_hE_QGSP_BIC_EMV.GetXaxis().GetBinUpEdge(
                            scaled_hist_hE_QGSP_BIC_EMV.FindLastBinAbove(
                                fit_selection_value,1,lower_peak_bin,scaled_hist_hE_QGSP_BIC_EMV.GetNbinsX())))
```
