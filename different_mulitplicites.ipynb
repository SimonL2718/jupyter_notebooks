```python
import ROOT
from ROOT import TFile, TCanvas, TLegend
```

```python
one_all_file_name="~/test/R3BRoot/build/bin/one_all.root"
two_all_file_name="~/test/R3BRoot/build/bin/two_all.root"
three_all_file_name="~/test/R3BRoot/build/bin/three_all.root"
four_all_file_name="~/test/R3BRoot/build/bin/four_all.root"
one_all_file=TFile(one_all_file_name,"read")
two_all_file=TFile(two_all_file_name,"read")
three_all_file=TFile(three_all_file_name,"read")
four_all_file=TFile(four_all_file_name,"read")
one_all_file_hist_monitor=one_all_file.GetDirectory("HistMonitor")
two_all_file_hist_monitor=two_all_file.GetDirectory("HistMonitor")
three_all_file_hist_monitor=three_all_file.GetDirectory("HistMonitor")
four_all_file_hist_monitor=four_all_file.GetDirectory("HistMonitor")
one_all_hit_energy_hist=one_all_file_hist_monitor.Get("hE")
two_all_hit_energy_hist=two_all_file_hist_monitor.Get("hE")
three_all_hit_energy_hist=three_all_file_hist_monitor.Get("hE")
four_all_hit_energy_hist=four_all_file_hist_monitor.Get("hE")

one_protons_file_name="~/test/R3BRoot/build/bin/one_protons.root"
two_protons_file_name="~/test/R3BRoot/build/bin/two_protons.root"
three_protons_file_name="~/test/R3BRoot/build/bin/three_protons.root"
four_protons_file_name="~/test/R3BRoot/build/bin/four_protons.root"
one_protons_file=TFile(one_protons_file_name,"read")
two_protons_file=TFile(two_protons_file_name,"read")
three_protons_file=TFile(three_protons_file_name,"read")
four_protons_file=TFile(four_protons_file_name,"read")
one_protons_file_hist_monitor=one_protons_file.GetDirectory("HistMonitor")
two_protons_file_hist_monitor=two_protons_file.GetDirectory("HistMonitor")
three_protons_file_hist_monitor=three_protons_file.GetDirectory("HistMonitor")
four_protons_file_hist_monitor=four_protons_file.GetDirectory("HistMonitor")
one_protons_hit_energy_hist=one_protons_file_hist_monitor.Get("hE")
two_protons_hit_energy_hist=two_protons_file_hist_monitor.Get("hE")
three_protons_hit_energy_hist=three_protons_file_hist_monitor.Get("hE")
four_protons_hit_energy_hist=four_protons_file_hist_monitor.Get("hE")

one_no_electrons_positrons_gamma_file_name="~/test/R3BRoot/build/bin/one_no_electrons_positrons_gamma.root"
two_no_electrons_positrons_gamma_file_name="~/test/R3BRoot/build/bin/two_no_electrons_positrons_gamma.root"
three_no_electrons_positrons_gamma_file_name="~/test/R3BRoot/build/bin/three_no_electrons_positrons_gamma.root"
four_no_electrons_positrons_gamma_file_name="~/test/R3BRoot/build/bin/four_no_electrons_positrons_gamma.root"
one_no_electrons_positrons_gamma_file=TFile(one_no_electrons_positrons_gamma_file_name,"read")
two_no_electrons_positrons_gamma_file=TFile(two_no_electrons_positrons_gamma_file_name,"read")
three_no_electrons_positrons_gamma_file=TFile(three_no_electrons_positrons_gamma_file_name,"read")
four_no_electrons_positrons_gamma_file=TFile(four_no_electrons_positrons_gamma_file_name,"read")
one_no_electrons_positrons_gamma_file_hist_monitor=one_no_electrons_positrons_gamma_file.GetDirectory("HistMonitor")
two_no_electrons_positrons_gamma_file_hist_monitor=two_no_electrons_positrons_gamma_file.GetDirectory("HistMonitor")
three_no_electrons_positrons_gamma_file_hist_monitor=three_no_electrons_positrons_gamma_file.GetDirectory("HistMonitor")
four_no_electrons_positrons_gamma_file_hist_monitor=four_no_electrons_positrons_gamma_file.GetDirectory("HistMonitor")
one_no_electrons_positrons_gamma_hit_energy_hist=one_no_electrons_positrons_gamma_file_hist_monitor.Get("hE")
two_no_electrons_positrons_gamma_hit_energy_hist=two_no_electrons_positrons_gamma_file_hist_monitor.Get("hE")
three_no_electrons_positrons_gamma_hit_energy_hist=three_no_electrons_positrons_gamma_file_hist_monitor.Get("hE")
four_no_electrons_positrons_gamma_hit_energy_hist=four_no_electrons_positrons_gamma_file_hist_monitor.Get("hE")
```

```python
%jsroot on
canvas=TCanvas("hE","Hit Energy",0,0,2000,2000)
canvas.Divide(2,2)
canvas.cd(1)
one_all_hit_energy_hist.SetTitle("n=1, E=1562.08MeV")

one_all_hit_energy_hist.Draw()
one_all_hit_energy_hist.SetLineColor(4)
one_protons_hit_energy_hist.Draw("same hist")
one_protons_hit_energy_hist.SetLineColor(2)
one_no_electrons_positrons_gamma_hit_energy_hist.Draw("same hist")
one_no_electrons_positrons_gamma_hit_energy_hist.SetLineColor(6)

one_all_hit_energy_hist.SetStats(0)
one_protons_hit_energy_hist.SetStats(0)
one_no_electrons_positrons_gamma_hit_energy_hist.SetStats(0)

one_legend=TLegend(0.6,0.7,0.9,0.9)
one_legend.SetHeader("Hit Energies", "c")
one_legend.AddEntry(one_all_hit_energy_hist,"all particles","l")
one_legend.AddEntry(one_protons_hit_energy_hist,"only protons","l")
one_legend.AddEntry(one_no_electrons_positrons_gamma_hit_energy_hist,"no e⁺,e⁻, γ","l")
one_legend.Draw()

one_all_hit_energy_hist.GetYaxis().SetRangeUser(0,3000)

canvas.cd(2)
two_all_hit_energy_hist.SetTitle("n=2, E=1562.08MeV")

two_all_hit_energy_hist.Draw()
two_all_hit_energy_hist.SetLineColor(4)
two_protons_hit_energy_hist.Draw("same hist")
two_protons_hit_energy_hist.SetLineColor(2)
two_no_electrons_positrons_gamma_hit_energy_hist.Draw("same hist")
two_no_electrons_positrons_gamma_hit_energy_hist.SetLineColor(6)

two_all_hit_energy_hist.SetStats(0)
two_protons_hit_energy_hist.SetStats(0)
two_no_electrons_positrons_gamma_hit_energy_hist.SetStats(0)

two_legend=TLegend(0.6,0.7,0.9,0.9)
two_legend.SetHeader("Hit Energies", "c")
two_legend.AddEntry(two_all_hit_energy_hist,"all particles","l")
two_legend.AddEntry(two_protons_hit_energy_hist,"only protons","l")
two_legend.AddEntry(two_no_electrons_positrons_gamma_hit_energy_hist,"no e⁺,e⁻,γ","l")
two_legend.Draw()

two_all_hit_energy_hist.GetYaxis().SetRangeUser(0,6000)

canvas.cd(3)
three_all_hit_energy_hist.SetTitle("n=3, E=1562.08MeV")

three_all_hit_energy_hist.Draw()
three_all_hit_energy_hist.SetLineColor(4)
three_protons_hit_energy_hist.Draw("same hist")
three_protons_hit_energy_hist.SetLineColor(2)
three_no_electrons_positrons_gamma_hit_energy_hist.Draw("same hist")
three_no_electrons_positrons_gamma_hit_energy_hist.SetLineColor(6)

three_all_hit_energy_hist.SetStats(0)
three_protons_hit_energy_hist.SetStats(0)
three_no_electrons_positrons_gamma_hit_energy_hist.SetStats(0)

three_legend=TLegend(0.6,0.7,0.9,0.9)
three_legend.SetHeader("Hit Energies", "c")
three_legend.AddEntry(three_all_hit_energy_hist,"all particles","l")
three_legend.AddEntry(three_protons_hit_energy_hist,"only protons","l")
three_legend.AddEntry(three_no_electrons_positrons_gamma_hit_energy_hist,"no e⁺,e⁻,γ","l")
three_legend.Draw()
three_all_hit_energy_hist.GetYaxis().SetRangeUser(0,9000)

canvas.cd(4)
four_all_hit_energy_hist.SetTitle("n=4, E=1562.08MeV")

four_all_hit_energy_hist.Draw()
four_all_hit_energy_hist.SetLineColor(4)
four_protons_hit_energy_hist.Draw("same hist")
four_protons_hit_energy_hist.SetLineColor(2)
four_no_electrons_positrons_gamma_hit_energy_hist.Draw("same hist")
four_no_electrons_positrons_gamma_hit_energy_hist.SetLineColor(6)

four_all_hit_energy_hist.SetStats(0)
four_protons_hit_energy_hist.SetStats(0)
four_no_electrons_positrons_gamma_hit_energy_hist.SetStats(0)

four_legend=TLegend(0.6,0.7,0.9,0.9)
four_legend.SetHeader("Hit Energies", "c")
four_legend.AddEntry(four_all_hit_energy_hist,"all particles","l")
four_legend.AddEntry(four_protons_hit_energy_hist,"only protons","l")
four_legend.AddEntry(four_no_electrons_positrons_gamma_hit_energy_hist,"no e⁺,e⁻,γ","l")
four_legend.Draw()

four_all_hit_energy_hist.GetYaxis().SetRangeUser(0,12000)

canvas.Draw()
canvas.SaveAs("mult.png")
```
